```python
#joints
import pandas as pd
ddpt={
    'deptid':[1,2,3,4,5,6],
    'depname':['cs','it','ec','me','el','ml']
}

dept=pd.DataFrame(ddpt)
dept

```

```python
employ={
    'eid':[i for i in range(100,106)],
    'ename':['sa','da','ram','pra','mok','par'],
    'edesignation':['hr','anala','manger','leader','tl','lead'],
    'deptidd':[2,4,5,8,7,2]
}

ep=pd.DataFrame(employ)

ep

```

```python
#joints

innerjoin=pd.merge(ep,dept,how='inner' , left_on='deptidd', right_on='deptid')

innerjoin
```
