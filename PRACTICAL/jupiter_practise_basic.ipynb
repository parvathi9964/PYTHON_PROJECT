```python
#deleting  dict

d={1:"r",2:"t",3:"rew",4:"tre"}
del d[4]

print(d)

a=d.pop(2)
print(a)

key,value=d.popitem()
print(key,value)

```

```python
#iterate in dict
d={1:"r",2:"t",3:"rew",4:"tre"}

for key in d:
    print(key)

for value in d.values():
    print(value)

for key,value in d.items():
    print(key,value)
```
