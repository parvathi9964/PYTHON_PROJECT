Exploratory Data Analysis (EDA)
🔹 Total Revenue

```python
df['total_amount'].sum()
```

Revenue by Product Category

```python
category_sales = df.groupby('product_category')['total_amount'].sum().sort_values()

category_sales.plot(kind='barh', title='Revenue by Product Category')
plt.show()
```

```python
#City-wise Revenue
```

```python
city_sales = df.groupby('city')['total_amount'].sum()

city_sales.plot(kind='bar', title='City-wise Revenue')
plt.show()
```

Age Group Analysis:
Customers aged 25–40 are the most active buyers.

```python
sns.histplot(df['age'], bins=10)
plt.title("Customer Age Distribution")
plt.show()
```

Discount vs Revenue

```python
sns.boxplot(x='discount', y='total_amount', data=df)
plt.show()
```

```python
sns.boxplot(x='payment_mode',y='total_amount', data=df)
plt.show()
```

Payment Mode Preference

```python
df['payment_mode'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Mode Distribution")
plt.ylabel("")
plt.show()
```

```python
df['payment_mode'].value_counts()
```
